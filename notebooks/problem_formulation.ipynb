# Our inference strategy

We have a set of lenses, each with a set of photometric observables $\vec{d}$ plus time delays $\Delta t$.
We choose a set of parameters $\vec{a}$ describing each lens. For a given value of $\vec{a}$ we can predict our photometric observables $\vec{d}$. The values of $\vec{a}$ are drawn from a distribution that we choose to describe in terms of a set of hyperparameters $\vec{\eta}$. In addition, we have a hyperparameter $h$ that affects the time delay observable $\Delta t$.
